## Pandas Filtering and Selection

Explored dataset filtering and selection using Pandas. Practiced conditional filtering, column selection, and subsetting techniques to extract and analyze specific portions of data efficiently.

```python
import pandas as pd
data= {
    'Gene_ID':['G1', 'G2', 'G3', 'G4','G5'],
    'Organism':['E.coli', 'S.aureus', 'B.subtilis', 'E.coli', 'S.aureus'],
    'Length_bp':[1500, 1350, 2100, 1750, 1980],
    'GC_Content':[52.3, 38.9, 44.1, 55.6, 35.4]
}

df= pd.DataFrame(data)

df.to_csv("gene_sequences.csv", index= False)
df= df.loc[(df['Organism']== 'E.coli') & (df['GC_Content'] > 50)]
print(df[['Gene_ID', 'GC_Content']])

```
