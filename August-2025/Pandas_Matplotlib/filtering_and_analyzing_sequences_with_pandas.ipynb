Question:

Using the same gene_sequences.csv data (or the df you already have), do the following:

1. Add a new column called AT_Content which is calculated as:
AT_Content = 100 - GC_Content


2. Filter the dataframe to show only rows where AT_Content is greater than 60%.


3. Print the Gene_ID, Organism, and AT_Content columns for those filtered rows.

```python
import pandas as pd

df= pd.read_csv("gene_sequences.csv")
df['AT_Content']= 100 - df['GC_Content']
df= df.loc[df['AT_Content'] > 60]
print(df[['Gene_ID', 'Organism', 'AT_Content']])
```

Question 2:
From the filtered dataframe in Question 2, group the data by Organism and calculate:

The average GC_Content for each organism

The total number of genes for each organism


Print the results clearly.

```python
import pandas as pd

df= pd.read_csv("gene_sequences.csv")
df['AT_Content']= 100 - df['GC_Content']
df= df.loc[df['AT_Content'] > 60]

avg_gc= df.groupby('Organism')['GC_Content'].mean()
total_no_of_genes= df.groupby('Organism')['Gene_ID'].count()
print(f"Average GC_Content/Organism: {avg_gc}\n")
print(f"Total number of genes/Organism: {total_no_of_genes}")
```
