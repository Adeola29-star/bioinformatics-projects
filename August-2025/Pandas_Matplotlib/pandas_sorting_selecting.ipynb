Question 2:
Using the same gene_sequences.csv file, write code to:

Select all rows where Length_bp is greater than or equal to 1800

Sort the results by GC_Content in descending order

Display only the Gene_ID, Length_bp, and GC_ContentÂ columns

```python
import pandas as pd
df= pd.read_csv("gene_sequences.csv")

df= df.loc[df['Length_bp'] >= 1800].sort_values(by='GC_Content', ascending= False)
print(df[['Gene_ID', 'Length_bp', 'GC_Content']])
```
