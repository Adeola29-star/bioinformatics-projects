Exercise 1:

Write a function that takes a DNA sequence (cleaned), scans it from position 0 stepping by 3 bases, finds ORFs (start with ATG, ends with one of the stop codons), and returns a list of tuples (start_index, end_index, orf_sequence).

Use the fasta parsing function from before to get sequences from a file, then call this ORF finder on each sequence and print the results.

```python
def fasta_parsing(practice3txt):
    fasta_dict= {}
    with open("practice3.txt", "r") as file:
        header= None
        sequence_lines= []
        for line in file:
            line=line.strip()
            if line.startswith('>'):
                if header:
                    fasta_dict[header]= "".join(sequence_lines)
                header= line[1:]
                sequence_lines= []
            else:
                sequence_lines.append(line)
        if header:
            fasta_dict[header]= "".join(sequence_lines)
    return fasta_dict


def clean(sequence):
    cleaned= "".join([base for base in sequence.upper() if base in 'ATGC'])
    return cleaned


def find_orfs(dna):
    i= 0
    start_codon= 'ATG'
    stop_codons= {'TAA', 'TGA', 'TAG'}
    orfs= []
    while i < len(dna)-2:
        codon= dna[i:i+3]
        if codon == start_codon:
            for j in range(i+3, len(dna)-2, 3):
                stop_codon= dna[j:j+3]
                if stop_codon in stop_codons:
                    orf= dna[i:j+3]
                    orfs.append((i+1, j+3, orf))
                    i= j+3 # i+=3 to find orf in initial orf(ovelapping orfs)
                    break
            else:
                i += 3
        else:
            i += 3
    return orfs
            

sequences= fasta_parsing("practice3.txt")

cleaned_seqs= {}
for header, seq in sequences.items():
    cleaned_seqs[header]= clean(seq)

for header, seq in cleaned_seqs.items():
    print(f"Header: {header}\nClean Sequence: {seq}\nLength of sequence: {len(seq)}\n")
    orfs= find_orfs(seq)
    if orfs:
        for s_index, e_index, orf in orfs:
            print(f"start index: {s_index}\nend index: {e_index}\norf sequence: {orf}\n")
    else:
        print("No ORFs found in this sequence")
   
        
```
