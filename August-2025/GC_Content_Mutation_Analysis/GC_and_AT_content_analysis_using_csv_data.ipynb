## GC and AT Content Analysis Using CSV Data 

Created a custom CSV dataset and used Pandas for reading, filtering, and calculating average nucleotide compositions. 

Computed the average GC content per organism and filtered organisms with AT content > 60%, identifying one organism meeting the criteria. 

Visualized results with Matplotlib through comparative bar charts showing how higher AT content corresponds to lower GC content, demonstrating inverse nucleotide relationships in the dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd 

df= pd.read_csv("gene_sequences.csv")
df['AT_Content']= 100 - df['GC_Content']
df= df.loc[df['AT_Content'] > 60]

avg_gc= df.groupby('Organism')['GC_Content'].mean()
print(avg_gc)

x= avg_gc.index
y= avg_gc.values

plt.bar(x, y, color= 'orange')
plt.title("Average GC Content per Organism")
plt.xlabel("Organisms")
plt.ylabel("Average GC Content(%)")
plt.xticks(x, rotation= 90)
plt.tight_layout()
plt.show()



```
