## GC Content and ORF Analysis Pipeline with Visualization 
This script integrates custom bioinformatics functions (from parsing to translation) with data storage in Pandas and visualization using Matplotlib. 

It analyzes GC content across multiple DNA samples and visualizes GC distribution and frequency.

```python
dna= """
>Human_sequence Homo_sapiens
ATGCTAGCTAGCTAACGNNNATGCTAGCTAGCTGAC
>Mouse_sequence Mus_musculus
TTGCGCGGATCGTAGCTAGCTAGCTAGCTAATGCXTA
>Plant_sequence Arabidopsis_thaliana
GCTAGCTAGCATCGATCGTATAGCTAGCTAGCNNN
>Bacteria_sequence Escherichia_coli
ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG
>Virus_sequence Influenza_A
ATGTTTGCTGACTTCTTGGAATGGATGAGTTT"""

with open("day3_sequences.fasta", "w") as file:
    file.write(dna)
```

```python

import pandas as pd
from day import clean, gc_whole_sequences, transcription, orf_finder, translation


def load(filepath):
    fasta_dict= {}
    with open(filepath, "r") as file:
        header= None
        seq_lines= []
        for line in file:
            line= line.strip()
            if line.startswith('>'):
                if header:
                    fasta_dict[header]= "".join(seq_lines)
                header_parts= line[1:].split()
                header= header_parts[1] if len(header_parts) > 1 else header_parts[0]
                seq_lines= []
            else:
                seq_lines.append(line)
        if header:
            fasta_dict[header]= "".join(seq_lines)
    return fasta_dict

sequences= load("day3_sequences.fasta")
cleaned_seq= {}
for header, seq in sequences.items():
    cleaned_seq[header]= clean(seq)
    
import pandas as pd
results_for_pandas= []
for header, seq in cleaned_seq.items():
    Org= header
    Seq= seq
    length = len(seq)
    gc = gc_whole_sequences(seq)
    rna= transcription(seq)
    _, longest_orf, _ = orf_finder(rna)
    protein= translation(longest_orf)

    results_for_pandas.append({
        'Organism': Org,
        'Sequence': Seq,
        'Length': length,
        'GC_Content': gc,
        'Longest_ORF': longest_orf
    })


```

```python
df= pd.DataFrame(results_for_pandas)
print(df)

df.to_csv("sequence_metrics.csv", index=False)
```

```python
df= df.loc[df['GC_Content'] <= 50]
print(df)
```

```python
import matplotlib.pyplot as plt

x= df['Organism']
y= df['GC_Content']
plt.figure(figsize = (8,5))
plt.bar(x,y, width= 0.5, color= '#6a4c93')
plt.title("GC Content for Organisms")
plt.xlabel("Organisms")
plt.ylabel("GC Content (%)")
plt.xticks(rotation= 45)
plt.tight_layout()
plt.show()
           
         
```

```python
import matplotlib.pyplot as plt

plt.hist(df['GC_Content'], bins= 5, color= 'purple', edgecolor= 'black')
plt.title("Distribution of GC Content")
plt.xlabel("GC Content (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
