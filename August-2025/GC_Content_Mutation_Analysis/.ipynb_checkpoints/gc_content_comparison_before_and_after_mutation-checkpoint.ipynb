Question 1:
Using your existing DataFrame df, create a bar chart that compares the GC content (%) of the original sequences vs the mutated sequences side by side for each sequence header.

X-axis: Sequence headers (Human_sequence, Mouse_sequence, Plant_sequence)

Y-axis: GC content percentage

Show two bars per header: one for original GC, one for mutated GC

```python

import matplotlib.pyplot as plt
import pandas as pd
from unique_mutation_and_gc_content_pandas import * 

df = pd.read_csv("unique_mutation.csv")
x= df['Header']
gc_original= df['GC_Content(%)']
gc_mutated= df['Mutated_GC_Content(%)']

width= 0.35
x_indexes= range(len(x))


plt.bar(x_indexes, gc_original, width=width, label= 'Original GC%')
plt.bar([i + width for i in x_indexes], gc_mutated, width=width, label= 'Mutated GC%')
plt.xticks([i + width/2 for i in x_indexes], x, rotation= 45)
plt.title("GC_Content Across the Original Sequences and Mutated Sequences")
plt.xlabel("Sequence Headers")
plt.ylabel("GC Content Percentage")
plt.legend()
plt.tight_layout()
plt.show()
```
