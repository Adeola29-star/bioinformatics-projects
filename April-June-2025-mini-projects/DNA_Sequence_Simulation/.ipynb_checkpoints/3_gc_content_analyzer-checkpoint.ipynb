### Mini Project : GC Content Analyzer

#### Objective  
To build a Python program that calculates the GC content of a DNA sequence ‚Äî both full and partial ‚Äî while cleaning the input and handling invalid characters.

#### Description  
This project is part of my beginner-to-intermediate bioinformatics learning path.  
The goal is to analyze DNA sequence quality through GC content ‚Äî the percentage of Guanine (G) and Cytosine (C) bases ‚Äî which is crucial for understanding sequence stability, gene density, and biological relevance.

The program:
1. Cleans DNA input  
2. Calculates GC content in:
   - Full DNA sequence  
   - User-specified subsection  
3. Returns both GC content values  
4. Handles invalid characters¬†and¬†errors



```python
def gc_content_analyzer():
# Request a user input
    user_input = input("Enter a dna sequence:")

# Cleaning the DNA    
    cleaned_full_dna = []  
    for base in user_input.upper():
        if base in "ATCG":
            cleaned_full_dna.append(base)
    full_dna_sequence = ''.join(cleaned_full_dna)

# To ask a user to input a specific start and end index
    start = int(input("Enter the start index for the gc content subsection analysis:"))
    end = int(input("Enter the start index for the gc content subsection analysis:"))
    user_specified_subsection= full_dna_sequence[start:end]

# GC content for the full dna sequence
    G = full_dna_sequence.count('G')
    C = full_dna_sequence.count('C')
    total = len(full_dna_sequence)
    GC_content_full_dna = (G + C)/total * 100 if total > 0 else 0 

# GC content for the user_specified subsection
    G_sub = user_specified_subsection.count('G')
    C_sub = user_specified_subsection.count('C')
    total_sub = len(user_specified_subsection)
    GC_content_sub = (G_sub + C_sub)/total_sub * 100 if total_sub > 0 else 0

    return f"""Full DNA sequence GC_Content:{GC_content_full_dna:.2f}%,
                Full DNA sequence: {full_dna_sequence},
                user_specified_subsection(index 2-10): {user_specified_subsection}
               user-specified subsection GC_content:{GC_content_sub:.2f}%
            """

print(gc_content_analyzer())  
```

##### üîç Summary

- GC content helps estimate DNA stability.
- The function works by counting the number of 'G' and 'C' bases.
- Result is rounded to 2 decimal places for clarity.
