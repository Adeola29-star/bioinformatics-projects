# Codon Splitter

*Objective:*  
Split RNA sequences into codons for downstream analysis.

*Description:*  
Splits a given RNA sequence into consecutive codons (triplets) to prepare for translation or frequency analysis.

*Skills Applied:*  
- Python programming  
- String slicing  
- Loops and sequences  
- RNA sequence handling

```python
# Write a code that takes a DNA string and splits it into codons(groups of 3 bases)

def codon_splitter(rna):
    codons= []
    for i in range(0,len(rna),3):
        codon = rna[i:i+3]
        if len(codon) ==3:
            codons.append(codon)
    return f"codons:{codons}"

rna= input("Enter a rna sequence:").upper()
print(codon_splitter(rna))         
    
```

```python
# Another method

def codon_splitter(rna):
    codons= [rna[i:i+3] for i in range(0, len(rna), 3) if len(rna[i:i+3])== 3]
    return f"codons:{codons}"

rna= input("Enter an rna sequence:").upper()
print(codon_splitter(rna))
```
