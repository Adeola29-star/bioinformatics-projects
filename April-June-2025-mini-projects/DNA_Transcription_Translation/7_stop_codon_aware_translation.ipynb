# Stop Codon Aware Translation

*Objective:*  
Translate RNA to amino acids while respecting stop codons.

*Description:*  
Stops translation when a stop codon is encountered to mimic natural protein synthesis termination.

*Skills Applied:*  
- Python programming  
- Loops and conditionals  
- Codon mapping

Challenge 1: Translate RNA to Protein Until First Stop Codon Write a function that:

    Takes an RNA sequence
    Splits it into codons
    Translates the codons into amino acids using a dictionary
    Stops translation when a stop codon is encountered
    Returns the resulting amino acid sequence as a list



```python
def translate_to_protein(rna):
    codon_map= {
        'AUG': 'M',
        'UUU': 'F', 'UUC': 'F',
        'UUA': 'L', 'UUG': 'L',
        'UAA': '*', 'UAG': '*', 'UGA': '*',
    }

    codons= [rna[i:i+3] for i in range(0, len(rna), 3) if len(rna[i:i+3]) ==3]
    
    aminoacids= []
    has_unknown= False
    
    for codon in codons:
        aa = codon_map.get(codon, '?')
        if aa == '*':
            break
        if aa == '?':
            has_unknown= True
        aminoacids.append(aa)

    protein= ''.join(aminoacids)

    result= f" Amino Acids(Protein):{protein}"
    
    if has_unknown:
        result += " Warning: Some codons could not be translated (marked as '?')"
    return result

rna= input("Enter an rna sequence:").upper()
print(translate_to_protein(rna))
        
```
