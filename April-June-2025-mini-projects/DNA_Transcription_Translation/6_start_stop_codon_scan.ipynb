# Start and Stop Codon Scan

*Objective:*  
Identify positions of start and stop codons in sequences.

*Description:*  
Scans RNA sequences to locate initiation and termination codons, useful for translation frame analysis.

CHALLENGE 1 : Identify start and stop codons in an RNA sequence

Concepts Tested:

- if/else
- string slicing
- basic logic flow
- bioinformatics: start (AUG) and stop codons (UAA,UAG,UGA)

  Task
  Write a function called check_start_stop(rna) that:
1. Takes an RNA string as input
2. Checks:
    - if the first codon is a start codon (AUG)
    - if the last codon is a stop codon (UAA, UAG, UGA)
3. Returns a string:
    - "Valid start and stop codons" if both conditions are true
    - "Start codon missing" if only stop is present.
    - "Both start and stop codons missing" if neither is present.



```python
def check_start_stop(rna):

    codon_map = codon_map= {
        'AUG': 'M',
        'UUC': 'F',
        'GCC': 'A',
        'UGG': 'W',
        'CCC': 'P',
        'AAA': 'K',
        'UUU': 'F',
        'UAA': 'Stop',
        'UAG': 'Stop',
        'UGA': 'Stop'
     }
    codons = [rna[i:i+3] for i in range(0, len(rna), 3) if len(rna[i:i+3])==3]
    if not codons:
        return "Both start and stop codons missing"

    start_codon = codons[0]
    stop_codon = codons[-1]

    is_start = start_codon == "AUG"
    is_stop = stop_codon in ["UAA", "UAG", "UGA"]

    if is_start and is_stop:
        return "Valid start and stop codons"
    elif is_start and not is_stop:
        return "Stop codon missing"
    elif not is_start and is_stop:
        return "Start codon missing"
    else:
        return "Both start and stop codons missing"

rna= input("Enter an rna sequence:").upper()
print(check_start_stop(rna))
```
