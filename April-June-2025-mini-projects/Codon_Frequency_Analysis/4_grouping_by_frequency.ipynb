# Grouping by Frequency

*Objective:*  
Group codons or amino acids based on their frequency ranges.

*Description:*  
Categorizes codons or amino acids into frequency bins to identify overrepresented or underrepresented groups.

*Skills Applied:*  
- Python programming  
- Dictionaries and loops  
- Frequency analysis  
- Data grouping


```python
# Group amino acids by frequency count

freq={'M':2,'F':4,'L':2}
grouped={}
for aa,count in freq.items():
    grouped.setdefault(count,[]).append(aa)
print("Grouped:",grouped)
```
