# Filtering Valid Codons

*Objective:*  
Identify and filter valid codons from RNA sequences.

*Description:*  
Checks RNA sequences and retains only valid codons, discarding incomplete or invalid triplets.

*Skills Applied:*  
- Python programming  
- Loops and conditionals  
- Sequence validation


```python
def filtering_for_valid_codons(rna):
    codon_map= {
        'AUG': 'M',
        'UUU': 'F', 'UUC': 'F',
        'UUA': 'L', 'UUG': 'L',
        'UAA': '*', 'UAG': '*', 'UGA': '*',
    }
    valid_bases= {'A', 'U','G', 'C'}

# Slice RNA into codons of 3 characters
    codons= [rna[i:i+3] for i in range(0, len(rna),3) if len(rna[i:i+3]) == 3]

# Use filter() and lambda to keep any valid codons
    valid_codons= list(filter(lambda codon: all(base in valid_bases for base in codon), codons))

    return f" Valid Codons: {valid_codons}"

rna= input("Enter an rna sequence:").upper()
print(filtering_for_valid_codons(rna))
    
```
