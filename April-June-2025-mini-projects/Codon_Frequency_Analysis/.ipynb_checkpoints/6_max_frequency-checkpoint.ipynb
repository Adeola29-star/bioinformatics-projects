# Maximum Frequency

*Objective:*  
Determine the codon or amino acid with the highest frequency in a sequence.

*Description:*  
Analyzes sequences and identifies the most frequently occurring codon or amino acid, useful for studying codon bias.

*Skills Applied:*  
- Python programming  
- Dictionaries and max functions  
- Sequence analysis


```python
def highest_aa(rna):
    codon_map = {
    'AUG': 'M',
    'UUU': 'F', 'UUC': 'F',
    'UUA': 'L', 'UUG': 'L',
    'UAA': '*', 'UAG': '*', 'UGA': '*',
    }

    codons= [rna[i:i+3] for i in range(0, len(rna), 3) if len(rna[i:i+3]) ==3]

    aminoacids= []
    for codon in codons:
        aa = codon_map.get(codon, '?')
        if aa == '*':
            break
        aminoacids.append(aa)

    frequency= {}
    for aa in aminoacids:
        frequency[aa]= frequency.get(aa, 0) + 1

    if not frequency:
        return "No amino acids before stop codon."

    most_frequent= max(frequency.items(), key=lambda item: item[1])
    return f"""most frequent amino acid:{most_frequent},
                Amino acid : {most_frequent[0]},
                count : {most_frequent[1]}"""    
                       
rna= input("Enter an rna sequence:").upper()
print(highest_aa(rna))
```
