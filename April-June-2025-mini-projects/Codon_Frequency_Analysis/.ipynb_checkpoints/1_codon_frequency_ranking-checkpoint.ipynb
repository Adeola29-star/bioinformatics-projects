# Codon Frequency and Ranking

*Objective:*  
Analyze codon usage frequency and rank codons by occurrence.

*Description:*  
Calculates how often each codon appears in a sequence and ranks them from most to least frequent. Useful in understanding codon bias.

*Skills Applied:*  
- Python programming  
- Dictionaries and sorting  
- Frequency analysis  
- Bioinformatics concepts

```python
def codon_frequency(rna):
    codons = [ ]
    for i in range(0, len(rna),3):
        codon = rna[i:i+3]
        if len(codon) == 3:
            codons.append(codon)
            
    codon_counts = { }
    for codon in codons:
        if codon in codon_counts:
            codon_counts[codon] +=1
        else:
            codon_counts[codon] = 1
            
    max_frequency= max(codon_counts, key=codon_counts.get)
    return max_frequency,codon_counts[max_frequency]

rna = input("Enter an rna sequence:").upper()

print(codon_frequency(rna))
```
