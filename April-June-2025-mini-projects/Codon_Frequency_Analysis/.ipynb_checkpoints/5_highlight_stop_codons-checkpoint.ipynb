# Highlight Stop Codons

*Objective:*  
Identify and mark stop codons in RNA or DNA sequences.

*Description:*  
Detects stop codons (UAA, UAG, UGA) in sequences and highlights their positions to assist translation analysis.

*Skills Applied:*  
- Python programming  
- String matching and searching  
- Codon analysis

```python
def highlight_stop_codons(rna):
     codon_map= {
        'AUG': 'M',
        'UUU': 'F', 'UUC': 'F',
        'UUA': 'L', 'UUG': 'L',
        'UAA': 'STOP', 'UAG': 'STOP', 'UGA': 'STOP',
    }
     codons=[rna[i:i+3] for i in range(0,len(rna),3) if len(rna[i:i+3])==3]
     result = []
     for codon in codons:
         if codon_map.get(codon) == 'STOP':
             result.append(f"[{codon}]")
         else:
             result.append(codon)
     
     return f" Highlighted codon : {result}"

rna= input("Enter an RNA sequence:").upper()
print(highlight_stop_codons(rna))
    
```

```python
def highlight_stop_codons(rna):
    codon_map= {
        'AUG': 'M',
        'UUU': 'F', 'UUC': 'F',
        'UUA': 'L', 'UUG': 'L',
        'UAA': 'STOP', 'UAG': 'STOP', 'UGA': 'STOP',
    }
    codons= [rna[i:i+3] for i in range(0, len(rna), 3) if len(rna[i:i+3])==3]
    result= []
    for codon in codons:
        if codon_map.get(codon) == 'STOP':
            result.append(f"{codon}*")
            
        elif codon_map.get(codon, '?') == '?':
            result.append('unknown')
        else:
            result.append(codon)
    return f"Highlighted Codons :{result}"

rna= input("Enter an rna sequence:").upper()
print(highlight_stop_codons(rna))
    
```
