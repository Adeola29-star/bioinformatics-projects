# TopN Frequency

*Objective:*  
Identify the top N most frequent codons or amino acids.

*Description:*  
Ranks codons or amino acids by frequency and extracts the top N most common entries for analysis.

*Skills Applied:*  
- Python programming  
- Sorting and ranking  
- Dictionaries and data analy

Find Amino Acid Frequencies and Return Top 3

concepts practiced: for loops, dictionary counting, sorted(), lambda functions, slicing and control structures, using .get() to safely fetch dictionary values

write a function that:

    Translates an RNA sequence into amino acids until the first stop codon (use your codon_map).
    Counts the frequency of each amino acid (excluding the stop codon)
    Returns the top 3 most frequent amino acids in the format: [(amino_acid, count), (amino_acid, count), .....]



```python
def frequency_aa(rna):
    codon_map= {
        'AUG': 'M',
        'UUU': 'F', 'UUC': 'F',
        'UUA': 'L', 'UUG': 'L',
        'UAA': '*', 'UAG': '*', 'UGA': '*',
    }

# Split into codons
    codons= [rna[i:i+3] for i in range(0, len(rna),3) if len(rna[i:i+3])==3]

# Translate into amino acids
    aminoacids= []
    has_unknown= False

    for codon in codons:
        aa= codon_map.get(codon, '?')
        if aa == '*':
            break
        if aa == '?':
            has_unknown= True
        aminoacids.append(aa)

# Count frequencies
    frequency= {}
    for aa in aminoacids:
        frequency[aa]= frequency.get(aa, 0) + 1

# Sort by count (descending) and get top 3    
    Top_3= sorted(frequency.items(), key= lambda item:item[1], reverse= True)[:3]
    return f" Top 3 most frequent Amino Acids : {Top_3}"

rna= input("Enter an rna sequence:").upper()
print(frequency_aa(rna))
    
```
