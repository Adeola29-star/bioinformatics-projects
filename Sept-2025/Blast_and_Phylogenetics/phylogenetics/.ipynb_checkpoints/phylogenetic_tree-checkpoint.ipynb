--- DRAWING A PHYLOGENETIC TREE ---

```python
# 1 -> Download your sequences online using Entrez and save them manually offline (Hybrid)

from Bio import Entrez, SeqIO

Entrez.email = "adeolasuleiman060@gmail.com"

ids = ["NP_000549.1", "NP_032223.1", "NP_990607.1", "NP_001081742.1", "NP_571013.1"]
handle = Entrez.efetch(db = "protein", id= ",".join(ids), rettype = "fasta", retmode = "text")
records = list(SeqIO.parse(handle, "fasta-blast"))
handle.close()


# 2 -> Rename or add extra info to each record before saving

species = ["Human", "Mouse", "Chicken", "Frog", "Zebrafish"]
for rec, specie in zip(records, species):
    rec.id = f"{specie}_{rec.id}"
    rec.description = f"{specie} {rec.description}"

with open("hemoglobin_alpha_renamed.fasta", "w") as f:
    SeqIO.write(records, f, "fasta")

print(f"Saved {len(records)} sequences to hemoglobin_alpha.fasta")

```

```python
# 3 -> Multiple Sequence Alignment

import subprocess

muscle_exe = "C:\\Users\\User\\Documents\\bioinformatics_tools\\muscle3.8.31_i86win32.exe"
input_file = "hemoglobin_alpha_renamed.fasta"
output_file = "Aligned_hemoglobin_alpha_renamed.fasta"

cmd= [muscle_exe, "-in", input_file, "-out", output_file]

subprocess.run(cmd, check = True)

print("MUSCLE alignment is complete!!")
```

```python
# 4 -> Parsing the aligned file/ Loading the alignment with AlignIO

from Bio import AlignIO

alignment = AlignIO.read("Aligned_hemoglobin_alpha_renamed.fasta", "fasta")

# 5 -> Make sure all sequences ID are unique
from collections import Counter
counts = Counter()
for record in alignment:
    counts[record.id] += 1
    if counts[record.id] > 1:
        record.id = f"{record.id}_{counts[record.id]}"

print(alignment)
print(f"Number of sequences: {len(alignment)}")
```

```python
# 5 -> Calculate the distance matrix

from Bio.Phylo.TreeConstruction import DistanceCalculator

# Use identity model (fraction of identical positions)
calculator = DistanceCalculator('identity')
dm = calculator.get_distance(alignment)

print(dm)
```

```python
# 6 -> Build the tree

from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
from Bio import Phylo

# Use Neighbor Joining (NJ) to build the tree
constructor = DistanceTreeConstructor()
tree = constructor.nj(dm)

# print ASCII tree in terminal
Phylo.draw_ascii(tree)
```

```python
# 7 -> Save the tree

Phylo.write(tree, "hemoglobin_alpha_renamed_tree.nwk", "newick")
```

--- GRAPHICAL VISUALIZATION OF THE TREE ---

```python

from Bio import Phylo
import matplotlib.pyplot as plt

tree = Phylo.read("hemoglobin_alpha_renamed_tree.nwk", "newick")

fig = plt.figure(figsize = (12,6))
axes = fig.add_subplot (1,1,1)

Phylo.draw(tree, axes = axes, do_show = False)
plt.title("Hemoglobin Alpha Phylogenetic Tree")
plt.show()

# Save as png for reports/presentations
fig.savefig("hemoglobin_tree.jpg", dpi=300)
```
