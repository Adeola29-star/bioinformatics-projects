# Mini-Project: Sequence Analysis Pipeline

### Overview

This mini-project implements a complete pipeline for analyzing raw DNA sequences from a FASTA file. Tasks include cleaning sequences, computing GC content, transcription, translation, ORF detection, and plotting GC distribution.

### Steps Implemented

- File Parsing

Parses a raw FASTA file and stores sequences in a dictionary


- Sequence Cleaning

Removes invalid characters

Converts sequences to uppercase


- GC Content Calculation

Calculates GC percentage for each sequence


- Transcription & Translation

Transcribes DNA → RNA

Translates RNA → Protein using a codon table

Stops translation at stop codons


- Finding Longest ORFs

Detects longest open reading frame in all 3 reading frames


- Data Storage

Stores results (GC%, RNA, Protein, ORF, Sequence length) in a Pandas DataFrame


- Visualization

Plots GC content distribution using Matplotlib


- Pairwise & Multiple Sequence Alignment

Pairwise alignment of two selected sequences (global, substitution matrix optional)

Multiple sequence alignment using MUSCLE for all sequences


### Outputs

DataFrame containing:

Organism name

Sequence length

GC content (%)

RNA sequence

Protein sequence

Longest ORF per frame


GC content distribution bar plot

Pairwise alignment results (printed)

Multiple sequence alignment saved as clean_align.fasta

```python
# 02-09-2025 Tuesday Mini-project

def file_parsing(filepath):
    fasta_dict= {}
    with open(filepath, "r") as file:
        header= None
        seq = ""
        for line in file:
            line= line.strip()
            if line.startswith(">"):
                if header:
                    fasta_dict[header]= seq
                header_parts= line[1:].split()
                header= header_parts[1] if len(header_parts) > 1 else header_parts[0]
                seq = "" 
            else:
                seq += line

        if header:
            fasta_dict[header] = seq
    return fasta_dict


def clean(seq):
    cleaned_seq= "".join([base for base in seq.upper() if base in "ATGC"])
    return cleaned_seq

sequences= file_parsing("raw_sequences.fasta")

cleaned_file= {}
for header, seq in sequences.items():
    cleaned_file[header] = clean(seq)

def find_orf(seq):
    start_codon= "ATG"
    stop_codons = ("TAG", "TAA", "TGA")
    longest_orf= {1:None, 2:None, 3:None}
    orfs= []

    for frame in range(3):
        i = frame
        max_orf= ""
        while i < len(seq)-2:
            codon= seq[i:i+3] 
            if codon == start_codon:
                for j in range(i+3, len(seq)-2, 3):
                    stop_codon= seq[j:j+3] 
                    if stop_codon in stop_codons:
                        orf = seq[i:j+3]
                        if len(orf) > len(max_orf):
                            max_orf= orf
                            longest_orf[frame + 1] = max_orf
                        orfs.append(orfs)
                        i= j+3
                        break
                else:
                    i += 3
            else:
                i += 3
    return longest_orf

def calculate_gc(seq):
    gc= 0
    for base in seq:
        if base in "GC":
            gc += 1
    gc_percent = round((gc/len(seq)) * 100, 2)
    return gc_percent


def transcription(seq):
    rna= ""
    for base in seq:
        if base == "T":
            rna += "U"
        else:
            rna += base

    return rna


def translation(seq):
    codon_table = {
    'UUU': 'Phenylalanine', 'UUC': 'Phenylalanine',
    'UUA': 'Leucine', 'UUG': 'Leucine',
    'CUU': 'Leucine', 'CUC': 'Leucine', 'CUA': 'Leucine', 'CUG': 'Leucine',
    'AUU': 'Isoleucine', 'AUC': 'Isoleucine', 'AUA': 'Isoleucine',
    'AUG': 'Methionine',  # Start codon
    'GUU': 'Valine', 'GUC': 'Valine', 'GUA': 'Valine', 'GUG': 'Valine',

    'UCU': 'Serine', 'UCC': 'Serine', 'UCA': 'Serine', 'UCG': 'Serine',
    'CCU': 'Proline', 'CCC': 'Proline', 'CCA': 'Proline', 'CCG': 'Proline',
    'ACU': 'Threonine', 'ACC': 'Threonine', 'ACA': 'Threonine', 'ACG': 'Threonine',
    'GCU': 'Alanine', 'GCC': 'Alanine', 'GCA': 'Alanine', 'GCG': 'Alanine',

    'UAU': 'Tyrosine', 'UAC': 'Tyrosine',
    'CAU': 'Histidine', 'CAC': 'Histidine',
    'CAA': 'Glutamine', 'CAG': 'Glutamine',
    'AAU': 'Asparagine', 'AAC': 'Asparagine',
    'AAA': 'Lysine', 'AAG': 'Lysine',
    'GAU': 'Aspartic acid', 'GAC': 'Aspartic acid',
    'GAA': 'Glutamic acid', 'GAG': 'Glutamic acid',

    'UGU': 'Cysteine', 'UGC': 'Cysteine',
    'UGG': 'Tryptophan',
    'CGU': 'Arginine', 'CGC': 'Arginine', 'CGA': 'Arginine', 'CGG': 'Arginine',
    'AGU': 'Serine', 'AGC': 'Serine',
    'AGA': 'Arginine', 'AGG': 'Arginine',
    'GGU': 'Glycine', 'GGC': 'Glycine', 'GGA': 'Glycine', 'GGG': 'Glycine',
    'UAG': 'Stop', 'UGA': 'Stop', 'UAA': 'Stop'  # stop codons
}
    codons= [seq[i:i+3] for i in range(0, len(seq)-2, 3)]
    aminoacids= []
    for codon in codons:
        aa = codon_table.get(codon, '?')
        if aa== 'Stop':
            break
        aminoacids.append(aa)
    protein= "".join(aminoacids)
    return protein

results_pandas= []
for header, seq in cleaned_file.items():
    results_pandas.append({
        "Organism name": header,
        "Sequence length": len(seq),
        "GC content(%)": calculate_gc(seq),
        "RNA sequence": transcription(seq),
        "Protein sequence": translation(seq),
        "Longest ORF per frame": find_orf(seq)
})

import pandas as pd
df= pd.DataFrame(results_pandas)
print(df)

import matplotlib.pyplot as plt
x= df["Organism name"]
y= df["GC content(%)"]

plt.bar(x, y, color= "violet", width= 0.5)
plt.xlabel("Organisms")
plt.ylabel("GC Content(%)")
plt.xticks(rotation= 45)
plt.tight_layout()
plt.show()
```

```python

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.Align import PairwiseAligner

# Picked two sequences from my cleaned sequences for Pairwise Alignment
sequences= []
for record in SeqIO.parse("clean_sequences_copy.fasta", "fasta"):
    sequences.append(record.seq)

seq1= sequences[0]
seq2= sequences[1]

Rna1 = Seq(sequences[0]).transcribe()
Rna2 = Seq(sequences[1]).transcribe()

Protein1 = Rna1.translate()
Protein2 = Rna2.translate()


aligner= PairwiseAligner()
aligner.mode= "global"
aligner.match_score= 3
aligner.mismatch_score= -1
aligner.open_gap_score= -1
aligner.extend_gap_score= -2

alignments= aligner.align(seq1, seq2)

for aln in alignments:
    print(aln, aln.score)
    break


# Used a substitution matrix 
from Bio.Align import substitution_matrices

matrix= substitution_matrices.load("BLOSUM62")
aligner= PairwiseAligner()
aligner.mode= "global"
aligner.substitution_matrix= matrix

alignments= aligner.align(Protein1, Protein2)

for aln in alignments:
    print(aln, aln.score)
    break
```
