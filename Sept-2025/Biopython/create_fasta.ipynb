# Writing a FASTA File (HBB sequences)

Description:

This script creates a FASTA file named globin.fasta containing two sequences:

1. HBB_mRNA – nucleotide sequence of hemoglobin beta mRNA


2. HBB_protein – corresponding protein sequence




Purpose:

To generate a sample FASTA file for downstream BioPython parsing and analysis.

Demonstrates how to write multi-sequence FASTA data using Python file handling.


Output:

globin.fasta file with the nucleotide and protein sequences in standard FASTA format.

```python
Data= """
>HBB_mRNA
ATGGTGCATCTGACTCCTGAGGAGAAGTCTGCCGTTACTGCCCTGTGGGGCAAGGTGAACGTGGATGAAGTTGGTGGTGAGGCCCTGGGCAG
>HBB_protein
MVHLTPEEKSAVTALWGKVNVDEVGGEALGRLLVVYPWTQRFFESFGDLSTPDAVMGNPKVKAHGKKVLGAFSDGLAHLDNLKGTFATLSELHCDKLHVDPENFRLLGNVLVCVLAHHFGKEFTPPVQAAYQKVVAGVANALAHKYH"""


with open("globin.fasta", "w") as file:
    file.write(Data)
```
