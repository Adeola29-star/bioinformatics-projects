## Supporting Data Files

These are the files used as inputs or outputs for Biopython scripts in this folder:

### Files and Description

#### - globin.fasta	
Example DNA and protein sequences for complement, reverse complement, and basic Seq operations.

#### - practice3.fasta	
Test sequences for transcription, translation, ORF finding, and GC content analysis.

#### - clean.fasta	
Cleaned FASTA sequences created after removing invalid characters; used for alignment and plotting.

#### - cleaned.fasta	
Another cleaned sequence file, used in MUSCLE multiple sequence alignment (MSA).

#### - Aligned.fasta	
Output of MUSCLE MSA on clean.fasta.

#### - Aligned2.fasta	
Output of MUSCLE MSA on cleaned.fasta.


### Notes:

These files are used as input for the Python scripts demonstrating sequence handling, MSA, transcription, translation, and plotting.

Files can be regenerated using the corresponding scripts if they are missing.
