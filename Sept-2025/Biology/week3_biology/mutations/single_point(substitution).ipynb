```python
# RANDOM (REPEATED)
# mutations can occur at a particular position/positions multiple times

import random

def single_point_repeated(seq, n):
    seq_list= list(seq)
    bases= ("A", "T", "C", "G")
    
    for i in range(n):
        # pos = int(input("Enter a position:")) if i wanted the mutation occuring 
        # at specific positions
        pos= random.randint(0, len(seq)-1)

# The random picking of position/ positions cannot outweigh the len(seq) and has 
# automatically converted the pos to 0-based numbering. 

        index= pos

        original_base = seq_list[index]

# Selection of the new bases
        possible_outcomes= []
        for base in bases:
            if base != original_base:
                possible_outcomes.append(base)
                
# Picking the new bases at random
        new_base= random.choice(possible_outcomes)
        seq_list[index] = new_base
        
    mutated_seq= "".join(seq_list)
    return seq, mutated_seq

seq= "ATTGCGCGCGCG"
n= int(input("Enter the mutations to occur:"))
# pos= int(input("Enter a position number:")) if i were to specify a position

print(single_point_repeated(seq, n))
    
```

```python
# RANDOM (UNIQUE)
# mutations can occur at a particular position once

import random

def single_point_unique(seq, n):
    seq_list = list(seq)
    bases = ("A", "T", "C", "G")
    mutated_positions = set()

    while len(mutated_positions) < n:
        pos = random.randint(0, len(seq)-1)

        if pos not in mutated_positions:
            original_base = seq_list[pos]

            possible_outcomes = []
            for base in bases:
                if base != original_base:
                    possible_outcomes.append(base)

            new_base = random.choice(possible_outcomes)
            seq_list[pos] = new_base
            mutated_positions.add(pos)

    mutated_seq = "".join(seq_list)
    return seq, mutated_seq


seq= "ATTGCGCGCGCG"
n= int(input("Enter the numbers of mutations to occur:"))
# pos= int(input("Enter a position number:")) if i were to specify a position

print(single_point_unique(seq, n))
```
