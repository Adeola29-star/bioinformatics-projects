--- INSERTION MUTATION ---

```python
import random

def insertion(seq, n):

# n is just like the number of mutations to occur just like substitution but here
# it's insertion
    
    seq_list = list(seq)
    bases = ("A", "T", "G", "C")
    for i in range(n):

# can decide to pick positions at random or specify positions
# pos= int(input("Enter the position of choice:"))
        pos = random.randint(0, len(seq))

# new base/bases are to be picked from the base so as not to insert all the bases(tuple) 
# at once
        new_base = random.choice(bases)

# the insert() do not require assigning to a variable just like; extend, append, 
# reverse, sort, remove.
        seq_list.insert(pos, new_base)
    
    mutated_seq = "".join(seq_list)
    return seq, mutated_seq

seq= "ATGTGTCCTGTG"
n = int(input("Enter the number of insertions:"))

print(insertion(seq, n))
    
    
```
