```python
import random
def monohybrid_cross(parent1, parent2, n):

    offspring= []
    for child in range(n):
        allele_1= random.choice(parent1)
        allele_2= random.choice(parent2)
        child_genotype= "".join(sorted([allele_1, allele_2]))
        offspring.append(child_genotype)
    return offspring

parent1= input("Enter genotype:")
parent2= input("Enter genotype:")
n= int(input("Enter the number of offspring/children:"))

children = monohybrid_cross(parent1, parent2, n)

for i, genotype in enumerate(children, 1):
    print(f"Child {i}: Genotype = {genotype}")

from collections import Counter

genotype_count = Counter(children)
for genotype, count in genotype_count.items():
    print(f"Genotype ratios({genotype}: {count})\n")

phenotype_count= {"Dominant": 0, "Recessive": 0}
for genotype in children:
    if "A" in genotype:
        phenotype_count["Dominant"] += 1
    else:
        phenotype_count["Recessive"] += 1


for phenotype, count in phenotype_count.items():
    print(f"Phenotype ratios({phenotype}: {count})\n")
                
    
        
```

```python
import random

def dihybrid_cross(parent1, parent2, n):
    offspring = []

    p1_geneA, p1_geneB = parent1[:2], parent1[2:]
    p2_geneA, p2_geneB = parent2[:2], parent2[2:]
    
    for child in range(n):
        alleleA1= random.choice(p1_geneA)
        alleleA2= random.choice(p2_geneA)
        alleleB1= random.choice(p1_geneB)
        alleleB2= random.choice(p2_geneB)
        child_genotype = "".join(sorted([alleleA1, alleleA2])) + "".join(sorted([alleleB1, alleleB2]))
                                                                                 
        offspring.append(child_genotype)
    return offspring

parent1= input("Enter the genotype e.g AABb:")
parent2 = input("Enter the genotype:")
n = int(input("Enter the number of offspring:"))

children = dihybrid_cross(parent1, parent2, n)

for i, genotype in enumerate(children , 1):
    print(f"Child {i} : Genotype = {genotype}")

from collections import Counter
genotype_count = Counter(children)
for genotype, count in genotype_count.items():
    print(f"Genotype ratios = {genotype} : {count}\n")

phenotype_count = {"Both Dominant": 0, "One Dominant":0, "Both Recessive": 0}
for genotype in children:
    geneA = genotype[:2]
    geneB = genotype[2:]

    has_A_dominant = "A" in geneA
    has_B_dominant = "B" in geneB

    if has_A_dominant and has_B_dominant:
        phenotype_count["Both Dominant"] += 1
    elif has_A_dominant or has_B_dominant:
        phenotype_count["One Dominant"] += 1
    else:
        phenotype_count["Both Recessive"] += 1


for phenotype, count in phenotype_count.items():
    print(f"Phenotype ratios = {phenotype} : {count}\n")
```
