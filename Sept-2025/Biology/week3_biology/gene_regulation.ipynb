--- GENE REGULATION ---

```python
def lac_operon(glucose, lactose):
    
    if glucose == "present" and lactose == "absent":
        return "Operon is OFF"
    elif glucose == "absent" and lactose == "present":
        return "Operon is ON, transcription is High"
    elif glucose == "present" and lactose == "present":
        return "Operon is ON, but transcription is Low"
    else:
        return "Operon is OFF, no food taken"

glucose = input("Enter the presence (e.g present/absent):").lower()
lactose = input("Enter the presence (e.g present/absent):").lower()

print(lac_operon(glucose, lactose))
```
