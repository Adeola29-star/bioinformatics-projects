```python
import pandas as pd

df= pd.read_csv("codon_frequency.csv").sort_values('frequency', ascending = False).reset_index(drop=True)
print(df)
```

```python
# Top 3 most frequent codons
top3 = df.sort_values('frequency', ascending= False).head(3)
print("\nTop 3 Most Frequent Codons:")
print(top3)
```

```python
# Filter frequency between 15 and 25

filtered= df.loc[(df['frequency'] >= 15) & (df['frequency'] <= 25)]
print("\nCodons with Frequency Between 15 and 25:")
print(filtered)
```

```python
# Plot the codon frequencies

import matplotlib.pyplot as plt

df.plot(x= 'codon', y= 'frequency', kind='bar', 
        legend= False, color= 'teal')
plt.title("Codon Frequencies")
plt.xticks(rotation =45)
plt.xlabel("Codons")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
