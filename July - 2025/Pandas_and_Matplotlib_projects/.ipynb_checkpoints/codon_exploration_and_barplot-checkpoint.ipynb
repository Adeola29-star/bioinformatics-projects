# Codon Usage Analysis with Pandas and Matplotlib
This notebook reads the codon–amino acid–frequency dataset into a Pandas DataFrame.
It performs grouping, aggregation (total and mean frequency by amino acid), and plots visualizations using Matplotlib.

```python
import pandas as pd
df=pd.read_csv("codon_frequency2.csv")
df.drop(index = 3, inplace = True)
print(df)
```

```python
print("Codon + Frequency:")
print(df[['codon', 'Frequency']])
```

```python
print(df['Aminoacid'])
```

```python
print("Codon Aminoacid Columns:")
print(df[['codon', 'Aminoacid']])
```

```python
low_df= df.loc[df['Frequency'] < 10]
print("Codon with Frequency below 10:")
print(low_df)
```

```python
methionine= df.loc[df['Aminoacid']=='M']
print("Codon coding for methionine:")
print(methionine)
```

```python
sorted_df = df.sort_values('Frequency', ascending= True)
print(sorted_df)
```

```python
# 2 least sorted codons
least_sorted= df.sort_values('Frequency', ascending= False). tail(2)
print(least_sorted)
```

```python
print(df.describe())
```

Grouping & Aggregation

```python
# Total frequency per amino acid

total_freq_by_aa= df.groupby('Aminoacid')['Frequency'].sum()
print("Total frequency by amino acid:")
print(total_freq_by_aa)
```

```python
average_freq_by_aa= df.groupby('Aminoacid')['Frequency'].mean()
sorted= average_freq_by_aa.sort_values(ascending=False)
print("Average frequency by amino acid:")
print(average_freq_by_aa)
print(sorted)
```

Step 6: Bar Plot of Codon Frequencies

```python
import matplotlib.pyplot as plt

total_freq_by_aa = df.groupby('Aminoacid')['Frequency'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
total_freq_by_aa.plot(kind= 'bar', color= 'skyblue')
plt.title('Total Codon Frequency by AminoAcid')
plt.xlabel('Amino Acid')
plt.xticks(rotation =0)
plt.ylabel('Total Frequency')
plt.tight_layout()
plt.show()
```

```python
sorted_df= df.sort_values('Frequency',ascending= False)
plt.figure(figsize=(8,4))
plt.bar(sorted_df['codon'], sorted_df['Frequency'], color= 'purple')
plt.title('Codon Frequency')
plt.xlabel('Codon')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
