# Codon Frequency Data Creation and File Handling
This notebook creates a dataset of codons and their frequencies, writes it to a file, and reads it back for verification.
It demonstrates basic data handling with Python file operations and introduces structured biological data representation.

```python
data = """codon,frequency
'ATG',30  
'TAA',12
'GCG',22   
'AAA',10  
'CGT',18
'TTT',25
'GGA',16"""

with open("codon_frequency.csv", "w") as file:
    file.write(data)
```

```python
with open("codon_frequency.csv", "r") as file:
    print(file.read())
```
