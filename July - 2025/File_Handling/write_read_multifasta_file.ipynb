# Reading a Multi-Fasta File

## Description
This notebook demonstrates:
- Reading sequences from a multi-fasta file
- Writing sequences to a new file
- Handling multiple DNA sequences efficiently

Focuses on parsing multi-sequence files commonly used in bioinformatics.

```python
with open("multi.fasta", "w") as ffile:
    ffile.writelines(f">seq1\nATGCGTA\n>seq2\nTTGCGTAA\n>seq3\nCCGTAAGG\n")

with open("multi.fasta", "r") as file:
    for line in file:
        line= line.strip()
        if line.startswith('>'):
            header= line[1:]
            print(header)
```
