# Writing and Reading Sequence Lines

## Description
This notebook shows how to:
- Write sequence lines from one file into another
- Read back the sequences
- Handle file I/O in bioinformatics contexts

Exercise 4

Open header.txt (the same file).

Create a new file named onlyseqs.txt.

Write only the sequence lines (not the headers) into onlyseqs.txt.

Each sequence should be on its own line in the new file.

```python
with open("header.txt", "r") as file:
    with open("onlyseqs.txt", "w") as wfile:
        for line in file:
            line= line.strip()
            if not line.startswith('>'):
                wfile.write(line.strip() + '\n')
 
```

```python
with open("onlyseqs.txt", "r") as rfile:
    print(rfile.read())
```
