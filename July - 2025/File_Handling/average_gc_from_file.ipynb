# Average GC_Content from File

## Description
This notebook demonstrates:
- Reading DNA sequences from a file
- Calculating the average GC content
- Writing the results to a new file

It covers basic Python file handling operations including reading, writing, and iterating over file contents.

```python
with open("multi.fasta", "r") as file:

    gc_results=[]
    gc_values= []
  
    for line in file:
        line= line.strip()
        if line.startswith('>'):
            header= line[1:]
        else:
            sequence= line.upper()
            gc = 0
            for base in sequence:
                if base in 'GC':
                    gc +=1
            gc_percent = round(gc/len(sequence) * 100,2) if len(sequence)> 0 else 0
            gc_results.append((header, gc_percent))
            gc_values.append(gc_percent)
                     

with open("gc_report.txt", "w") as wfile:
    for header, gc_percent in gc_results:
        wfile.writelines(f"{header}\n{gc_percent}\n")
    average_gc= round(sum(gc_values)/len(gc_values), 2)
    print(f"Average GC Content: {average_gc}%")
```
