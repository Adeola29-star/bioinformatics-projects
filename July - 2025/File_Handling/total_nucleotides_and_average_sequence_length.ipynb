# Total Nucleotides and Average Sequence Length

## Description
This notebook demonstrates:
- Calculating the total number of nucleotides in a file
- Computing average length of sequences
- Applying Python loops and basic statistics on sequence data

Exercise 5

Open onlyseqs.txt (the file you just created).

Count and print:

The total number of nucleotides (i.e., total characters in all sequences, excluding newlines)

The average length of the sequences in that file (round to nearest whole number)

```python
total_nucleotides= 0
with open("onlyseqs.txt", "r") as file:
    lines= file.readlines()
    for line in lines:
        line= line.strip()
        total_nucleotides += len(line)
    number_of_sequences = len(lines)
    average_length = round(total_nucleotides/number_of_sequences)

    print(total_nucleotides)
    print(number_of_sequences)
    print(average_length)
        
```
