# Writing and Reading a File

## Description
This notebook covers:
- Writing data to a file
- Reading the file contents
- Using file I/O to store and retrieve DNA or RNA seque

```python
DNA= """
AAAAATTTTGGCGCGCTTTTAAATGTTGTCCTA
GGGTTTTTCTCTCTTCTAAAAAATTTTTTCGGGCGCGC
CTTTTTCCCCTTTATATAGATAGTAGATGCCGCTCGCCC
"""
with open("sample.txt", "w") as file:
    file.write(DNA)
```

```python
with open("sample.txt", "r") as file:
    print(file.read())
```
