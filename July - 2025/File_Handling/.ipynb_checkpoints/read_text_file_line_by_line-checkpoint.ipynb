Exercise 1: Reading a text file line by line

Create a text file named sample.txt (manually or by code).
Inside, put any 3–5 short DNA sequences (one per line).

Then, write Python code to:

Open sample.txt and

Print out each line exactly as it appears (including newline, if you wish).

```python
with open("sample.txt", "r") as dna:
    for line in dna:
        line= line.strip()   # strip() removes newline characters
        print(line)
```
